### Week 3
##### Data Modelling

```python
data.shape
```

```python
X = data.drop('Outcome',axis=1).values
y = data.Outcome.values
```

###### We will split our data into train and test set, therefore we'll use Scikit-learn because it provides the capability to split data into train and test using the train_test_split. We will then use a set size of 40% of the dataset.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neighbors = np.arange(1,9)
train_accuracy =np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
```

```python
for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
```

```python
plt.title('kNN Varying number of neighbors')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = knn.predict(X_test)
confusion_matrix(y_test,y_pred)
```

```python
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```

##### ROC curve

```python
y_pred_prob = knn.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr, label='Knn')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('Knn(n_neighbors=7) ROC curve')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_pred_prob)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'n_neighbors':np.arange(1,50)}
```

```python
knn = KNeighborsClassifier()
knn_cv= GridSearchCV(knn,param_grid,cv=5)
knn_cv.fit(X,y)
```

```python
knn_cv.best_score_
```
